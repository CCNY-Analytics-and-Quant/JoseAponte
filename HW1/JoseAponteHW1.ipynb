```python
"""
Homework 1
Author: Jose Aponte


"""

########## Functions ########## 

#LiabilitesAnnual calculates the annual liabilities
def liabilitiesAnnual():
    print("----- Insert Monthly Expenses: -----")
    print("Food")
    foodMonthly = userInput()
    print("Monthly Food Expenses = $",foodMonthly)
    print("Utilities")
    utilitesMonthly = userInput()
    print("Monthly Utilites/Bill Expenses = $",utilitesMonthly)
    print("Transport")
    transportMonthly = userInput()
    print("Monthy Transport Expenses = $", transportMonthly)
    print("Entertainment")
    entertainmentMonthly = userInput()
    print("Monthly Entertainment Expenses = $", entertainmentMonthly )
    print("Clothing")
    clothingMonthly = userInput()
    print("Monthly Clothing Expenses = $" , clothingMonthly)
    #print("Travel")
    #food = userInputExpenses()
    print("Medical")
    medicalMonthly = userInput()
    print("Monthly Medical Expenses = $", medicalMonthly)

    foodYearly = userInputYearly(foodMonthly)
    utilitesYearly = userInputYearly(utilitesMonthly)
    transportYearly = userInputYearly(transportMonthly)
    entertainmentYearly = userInputYearly(entertainmentMonthly)
    clothingYearly = userInputYearly(clothingMonthly)
    medicalYearly = userInputYearly(medicalMonthly)

    return foodYearly + utilitesYearly + transportYearly + entertainmentYearly + clothingYearly + medicalYearly
#userInput validates de inputs
def userInput():
    while True:
        value = float(input())
        if value > 0:
            return value
            break
        else:
            print("Value Must be Higher than Zero")

def userInputYearly(value):
    yearlyValue = value * 12
    return yearlyValue
    
#cashflowAnnual calculates the annual cashflow
def cashflowAnnual():
    print("----- CashFlow: -----")
    print("Annual Salary")
    salaryAnnually = userInput()
    print("Annual Investment")
    return salaryAnnually

def investment():
    print("----- Investment: -----")
    print("Are You Going To Invest:")
    print("(y) If Yes ")
    print("Anything Else if Not")
    selInvestment = input()
    selInvestment = selInvestment.lower()
    if selInvestment == "y":
        print("Starting Investment")
        startingAmount = float(input())
        print("Annual Contribution")
        annualContribution = float(input())
        print("Rate of Return (in %) ")
        rateOfReturn = float(input())
        rateOfReturnPercentage = rateOfReturn / 100
        print("Years to Grow")
        yearsToGrow = int(input())
        endBalance = (startingAmount*rateOfReturnPercentage) + annualContribution + startingAmount
        for i in range (2, yearsToGrow+1):
            endBalance = (endBalance*rateOfReturnPercentage) + annualContribution + endBalance
        return endBalance
    else:
        print("Without Investment")
        endBalance = 0
        return endBalance

def howMuchWillIhaveAtRetirementAge():
    print("How Much Will I Have at Retirement Age? Selected")
    print("What is Your Current Age?")
    currentAge = int(input())
    print(currentAge,"years old")
    print("What is Your Desired Retirement Age?")
    retirementAge = int(input())
    print(retirementAge, "years old")
    liabilitYearly = liabilitiesAnnual()
    cashAnnual = cashflowAnnual()
    totalInvesment = investment()
    print("Expected Increases In Salary (in %)")
    expectedIncreasesSalary = float(input())
    expectedIncreasesSalary = expectedIncreasesSalary/100
    endOfYear = 0
    print("Age\tSaves")
    for everyYear in range(currentAge, retirementAge+1):
        endOfYear += (cashAnnual-liabilitYearly)
        print(everyYear,"\t$" ,endOfYear)
        cashAnnual += (cashAnnual*expectedIncreasesSalary)
    lifetimeSaving = endOfYear + totalInvesment
    print("\nAge of retirement:",retirementAge,"years old")
    print("Amount of lifetime investment: $",totalInvesment)
    print("Total amount saved: $",endOfYear)

def whatAgeCanIRetireAt():
    print("What Age Can I Retire At? Selected")
    print("What is Your Current Age?")
    currentAge = int(input())
    print(currentAge,"years old")
    print("How Much Money Do You Want To Be Retired?")
    targetedSavings = float(input())
    print(targetedSavings, "is the goal for retirement")
    liabilitYearly = liabilitiesAnnual()
    cashYearly = cashflowAnnual()
    overallSavings = 0
    targetAge = currentAge
    totalInvestment = investment()
    print("Age\tSaves")
    while(overallSavings < targetedSavings):
        overallSavings += (cashYearly - liabilitYearly)
        print(targetAge,"\t$", overallSavings)
        targetAge += 1
    lifetimeSaving = overallSavings + totalInvestment
    print("\nAge of retirement:",targetAge-1,"years old")
    print("Amount of lifetime investment: $",totalInvestment)
    print("Total amount saved: $",lifetimeSaving)
        
########## Main ########## 

print("********* HW 1 *********\n")
while True:
    print("---------- Select an option ---------")
    print("(a)\tHow Much Will I Have at Retirement Age?")
    print("(b)\tWhat Age Can I Retire At?")
    print("Anything Else to Exit")
    selection = input()
    selectionLower = selection.lower()
    if selectionLower == "a":
        howMuchWillIhaveAtRetirementAge()
    elif selectionLower == 'b':
        whatAgeCanIRetireAt()
    else:
        print("\n\nScript terminated...")
        break
```
